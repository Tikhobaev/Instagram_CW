## 1. Log in

```python
from InstagramAPI import InstagramAPI
username = input()
password = input()
api = InstagramAPI(username, password)
api.login()
```

## 2. Get user_info and user_id

```python
user_name_for_parse = 'ilya_fisher_volga'
api.searchUsername(user_name_for_parse)
userinfo = api.LastJson
user_id = userinfo['user']['pk']
user_id
```

## 3. Get all user followers

```python
api.getUserFollowers(user_id)
followers_info = api.LastJson
followers = [[user['username'], user['pk']] for user in followers_info['users']]
counter = 1
for follower in followers:
    print(f'{counter})', follower[0], follower[1])
    counter += 1
```

## 4. Get all posts

```python
import time
posts = []
has_more_posts = True
max_id = ""
while has_more_posts:
    api.getUserFeed(user_id, maxid=max_id)
    if api.LastJson['more_available'] is not True:
        has_more_posts = False  # stop condition
        print("stopped")

    max_id = api.LastJson.get('next_max_id', '')
    posts.extend(api.LastJson['items'])  # merge lists
    time.sleep(2)  # Slows the script down to avoid flooding the servers
posts
```

## 5. Parse post names (sorted by likes count)

```python
top_posts = []

for post in posts:
    top_posts.append([int(post['like_count']), post['caption']['text'] if post['caption'] else 'Empty text'])
top_posts.sort(key = lambda x: x[0], reverse=True)
counter = 0
for post in top_posts:
    print('\n', '*****'*5, 'Post number', counter + 1, '*****'*5) 
    print('Number of likes =', post[0])
    print('Text:', post[1])
    counter += 1
```

## 6. Get urls of all pics in posts

```python
urls = []
for post in posts:
    try:
        for picture in post['image_versions2']['candidates']:
            urls.append(picture['url'])
            #print(picture['url'])
    except KeyError as e:
        for picture in post['carousel_media'][0]['image_versions2']['candidates']:
            urls.append(picture['url'])
for url in urls:
    print('\n', url, '\n')
```

## 7. Seach by hashtag

```python
import time
api.tagFeed('короновируситалия')
search_result = api.LastJson
search_result
```

## 8. Get all post with this hashtag 
##     sort it by number of likes
##     get media_ids

```python
overall_number_of_posts = len(search_result['ranked_items'])
posts = [post for post in search_result['ranked_items']]
media_ids = [post['id'] for post in posts]
print('Number of posts with this hashtag =', overall_number_of_posts)
top_posts = []

for post in posts:
    top_posts.append([int(post['like_count']), post['caption']['text'] if post['caption'] else 'Empty text'])
top_posts.sort(key = lambda x: x[0], reverse=True)
counter = 0
for post in top_posts:
    print('\n', '*****'*5, 'Post number', counter + 1, '*****'*5) 
    print('Number of likes =', post[0])
    print('Text:', post[1])
    counter += 1
```

## 9. Get likers info for every post with this hashtag
##     parse their ids and usernames

```python
hashtag_posts_likers = []
hashtag_posts_likers_ids_and_names = []
for media_id in media_ids:
    api.getMediaLikers(media_id)
    hashtag_posts_likers.append(api.LastJson)
    hashtag_posts_likers_ids_and_names.append(
        list([[user['pk'], user['username']] for user in api.LastJson['users']]))
print('User ids and usernames for the firts post')
hashtag_posts_likers_ids_and_names[0]
```
